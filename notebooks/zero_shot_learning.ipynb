```python
from transformers import pipeline
labels = ['Politics & Government','Business & Finance' , 'Science & Mathematics' ,'Society & Culture','Health']
pipe = pipeline(task='zero-shot-classification' , model="facebook/bart-large-mnli" )
```

```python
import pandas as pd
df = pd.read_csv('/content/df.csv' , usecols=['User','Tweets','Country'])
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
data = df.dropna()
```

```python
data['Tweets'].apply(lambda x : len(x)).describe()
```

```python
from tqdm.notebook import tqdm
features = []
for i in tqdm(data['Tweets']):
  x = pipe(i , labels,multi_label=True)
  features.append(x['labels'][0])
```

```python
feature=pd.DataFrame(features)
feature.to_csv('feature.csv')
```

```python
data['category'] = features
```
